```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import sklearn.tree as tree
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Cabin'] = df['Cabin'].fillna(method='pad')
df['Cabin'] = df['Cabin'].fillna(method='bfill')
df['Embarked'] = df['Embarked'].fillna(method='pad')
```

```python
df.isnull().sum()
```

```python
# dropign un-nessery columns
df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df.info()
```

```python
labelEncoder = LabelEncoder()
df['Sex'] = labelEncoder.fit_transform(df['Sex'])
df['Cabin'] = labelEncoder.fit_transform(df['Cabin'])
df['Embarked'] = labelEncoder.fit_transform(df['Embarked'])
```

```python
randomForestClassifier = RandomForestClassifier(n_estimators=1000,max_features=8,oob_score=True)
```

```python
randomForestClassifier.fit(df.drop(['Survived'],axis=1),df['Survived'])
```

```python
randomForestClassifier.oob_score_
```

```python
randomForestClassifier.feature_importances_
```

```python
for feature,importance in zip(df.columns,randomForestClassifier.feature_importances_):
    print(feature,'-',importance)
```

# Decision Tree on Age, Gender and Fare

```python
decisionTreeClassifier = DecisionTreeClassifier(max_depth=6)
```

```python
decisionTreeClassifier.fit(df[['Age','Sex','Fare']],df['Survived'])
```

```python
plt.rcParams["figure.figsize"] = (300,100)
```

```python
tree.plot_tree(decisionTreeClassifier,filled = True);
```
